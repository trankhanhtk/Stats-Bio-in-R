```r
set.seed(7654098)
n_samples = 1000
mx5 <- rep(NA, n_samples)
mx10=mx5
mx20=mx5
mx100=mx5

x_space = 1:6
for (i in 1:n_samples) {
    mx5[i] = mean(sample(x_space, 5, replace=TRUE))
    mx10[i] = mean(sample(x_space, 10, replace=TRUE))
    mx20[i] = mean(sample(x_space, 20, replace=TRUE))
    mx100[i] = mean(sample(x_space, 100, replace=TRUE))
}
```

```r
par(mfrow=c(2, 2), mar=c(6, 4, 0, 0), oma=c(4, 0, 0, 0))
hist(mx5, pch=19, xlim=c(1, 6), breaks=15, main='', col.axis='Peru', col='Gold', ylab='Density', xlab='Mean of 5 die rolls')
hist(mx10, pch=19, xlim=c(1, 6),breaks=15, main='', col.axis='Peru', col='Gold', ylab='', xlab='Mean of 10 die rolls')
hist(mx20, pch=19, xlim=c(1, 6),breaks=15, main='', col.axis='Peru', col='Gold', ylab='Density', xlab='Mean of 20 die rolls')
hist(mx100, pch=19, xlim=c(1, 6),breaks=15, main='', col.axis='Peru', col='Gold', ylab='', xlab='Mean of 100 die rolls')
```
