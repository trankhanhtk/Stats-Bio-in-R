## Expectation

### Descrete RV: Six-sided Die Rolling

```r
x = 1:6
p_x = 1/6
exp_x = sum(x * p_x)
exp_x
exp_x2 = sum(x^2 * p_x)
exp_x2
var_x = exp_x2 - exp_x^2
round(var_x, digits=2)
```

### Continuous RV: Uniform Distribution

```r
y1 <- runif(10000) #random samples from a U[0,1]
mean(y1)
y2 <- runif(10000,-2,15)
mean(y2)
```

## Density Plots

```r
par(mfrow = c(2, 2), oma=c(1,1,1,1))
x = seq(-3.5,3.5,length=101)
y <- dnorm(x)
plot(x, y, type='l', col='blue', ylab='Density', xlab='N(0,1)')
lines(c(0,0), c(0, .4), col='red')

x=seq(0,50,length=101)
y <- dexp(x, rate=1/5)
plot(x, y, type='l', col='blue', xlab='exp(1/5)')
lines(c(5, 5), c(0, .2), col='red')

x=seq(0,50,length=101)
y <- dgamma(x, shape=2, scale=5)
plot(x, y, type='l', col='blue', ylab='Density', xlab='Gamma(2, 5)')
lines(c(10, 10), c(0, .075), col='red')

x=seq(0, 1,length=101)
y <- dbeta(x, shape1=10, shape2=2)
plot(x, y, type='l', col='blue', xlab='Beta(10, 2)')
lines(c(0.833333333, 0.833333333), c(0, 4.3), col='red')
```
