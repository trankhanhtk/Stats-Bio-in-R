## Descrete IID RV

### Binomial Random Sampling

```r
observations = 20
p = .5
x = rbinom(observations, 1, p)
hist(x, breaks=20, col='blue', density=20)
```

```r
observations = 10000
p = .5
x = rbinom(observations, 1, p)
hist(x, breaks=20, col='blue', density=20)
```

```r
observations = 100000
p = .3
x = rbinom(observations, 1, p)
h <- hist(x, breaks=20, plot=F)
h$counts=h$counts/sum(h$counts)
plot(h, col='blue', ylim = c(0, 1), yaxt="n", xlab='Outcomes', ylab='Relative frequency')
axis(2, at = seq(0, 1, by=.1), las=1)
segments(0, p, 1, p, lwd=2, lty=3, col='orange')
```
