## Exponential Distribution

### Cumulative Probability Distribution

```r
surv_prob=pexp(6, 1/5, lower.tail = FALSE)
round(surv_prob,digits=3)
```

### Probability Density Distribution

```r
x_dexp <- seq(0, 20, by = 0.1)
length(x_dexp)
```

```r
x = dexp(x_dexp, rate = 1/5, log = FALSE)
```

```r
surv_pdf_eval <- c(seq(6,20,by=0.1),seq(20,6,by=-0.1))
```

```r
poly_surv <- c(rep(0,length(seq(6,20,by=0.1))), dexp(seq(20,6,by=-0.1), 1/5))
```

```r
plot(
    x_dexp, x, type="l", col="blue",lwd=3, xlab="Survival time (years)",
    ylab="Density", cex.lab=1.3,cex.axis=1.3,col.axis="orange"
)

polygon(surv_pdf_eval, poly_surv, col=rgb(0, 0, 1, 0.4),border=rgb(0, 0, 1, 0.4))
arrows(15,0.1,10,0.03,col="red",lwd=2,length = 0.15, angle = 20)
text(15,0.13, "Survival probability for 6+ years")
text(15,0.11,"0.301 = area under the pdf for x>6")
```

```r
surv_prob=pexp(6, 1/5)-pexp(5, 1/5)
round(surv_prob,digits=3)
```

```r
x_dexp_1 = seq(0, 20, by=.1)
x_1 = dexp(x_dexp_1, 1/5)
plot(x_dexp_1, x_1, type='l', xlab='Survival time (years)', ylab='Density', col='blue', cex.axis=1, cex.lab=1.3, col.axis='orange')
x_coord = c(seq(5, 6, by=.1), seq(6, 5, by=-.1))
y_coord = c(rep(0, length(seq(5, 6, by=.1))), dexp(seq(6, 5, by=-.1), 1/5))
polygon(x_coord, y_coord, col=rgb(0, 0, 1, 0.4),border=rgb(0, 0, 1, 0.4))
arrows(15, .15, 5.7, .07, col='red', lwd=2, angle=20)
text(13, .175, 'Survival probability for 5 to 6 years')
text(13, .16, '.067 = area under the PDF for X is in (5, 6)')
```

### Laplace Distribution

```r
library(VGAM)
#mean parameter
mu=2
#scale parameter
sigma=2
```

```r
y=rlaplace(10000,mu,sigma)
```

```r
hist(y,breaks=100,probability=TRUE,col=rgb(0,0,1,1/4),
     xlab="error",main="",cex.lab=1.3,cex.axis=1.3, col.axis="blue",ylim=c(0,.25))
x=seq(mu-15, mu+15,length=10001)
fx=dlaplace(x, location = mu, scale = sigma)
lines(x,fx,col="red",lwd=1)
```
