### Birthday Problem

```r
n=1:364 # Vector of number of people in the room
pn=n
# check tail and head
head(pn)
tail(pn)

#Prob. of at least 2 people ( >= 2 people) having the same B-day

for (i in 1:364) {
    pn[i]<- 1-prod(365:(365-i+1)/365)
}
```

```r
length(n)
length(pn)
? par
```

```r
plot(n, pn, type='o', col=c('blue', 'orange'), pch=19, col.axis='blue', lwd=.5, cex=.5, xlab='Number of people in room', ylab='P(at least 2 having same B-day)')
```

### Monte Carlo Simulation

```r
set.seed(7654098)
n=23 #number of people in the room
n_sim=10000 #number of Monte Carlo simulated rooms with n people
are_2_birthdays<-rep(NA,n_sim) #vector of indicators for >=2 same B-day

length(are_2_birthdays)
are_2_birthdays[0:10]
```

```r
for (i in 1:n_sim) {
    n_birthdays <- length(unique(sample(1:365, n, replace = TRUE)))
    are_2_birthdays[i]<-(n_birthdays<n)
}

mean(are_2_birthdays)
```
