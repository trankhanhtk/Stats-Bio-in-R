## Random Sampling

```r
## Generate 1000 observations from a Normal with mean 2 and variance 1
set.seed(33)
n_observations = 1000
x <- rnorm(n_observations, mean = 2, sd = 1)
```

## White Noise Plotting

```r
options(repr.plot.width=6, repr.plot.height=4)
plot(1:n_observations, x, pch=20, xlab='No. of Observations', ylab='Random sample from N(2, 1)', col='Blue', col.axis='orange')
```

## Two Ways of Plotting Q-Q Plot

```r
qqplot(qnorm(ppoints(n_observations, a = 0)), x, col='blue', pch=20, col.axis='orange', main="Q-Q plot for N(2, 1)")
```

```r
qqnorm(x, col='blue', pch=20, col.axis='orange', main="Q-Q plot for N(2, 1)")
```

## Likelihood vs Log Likelihood Function

```r
logL <- function(data, mu){
    vec <- -1*(data-mu)^2/(2*1)
    result <- sum(vec)
    
    return(result)
}
```

```r
logL(3, 0)
```

```r
L <- function(data, mu){
    vec <- dnorm(x=data, mean = mu, sd = 1)
    result <- prod(vec)
    return(result)
}
```

```r
round(L(data=x[1:3], mu=2), digits=4)
```
