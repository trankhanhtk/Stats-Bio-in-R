## Bernoulli Random Variables & Sampling

```r
n_trials = 10
```

```r
for (i in 1:n_trials) {
    x1 <- rbinom(21, 1, .5)
    cat(x1, "\n\n")
}
```

```r
binom_ps = seq(0, 1, length=21)

for (i in 1:n_trials) {
    x2 <- rbinom(21, 1, binom_ps)
    cat(x2, "\n\n")
}
```

## PDF of Poisson Random Variable

```r
poisson_pdf = function (lambda, x) {
    return(lambda^x * exp(1)^(-lambda) / factorial(x))
}
```

```r
poisson_pdf(20, 19) + poisson_pdf(20, 20)
```

```r
# P(X is in {18, 19, 20, 21})
poisson_pdf(20, 18) + poisson_pdf(20, 19) + poisson_pdf(20, 20) + poisson_pdf(20, 21)
```

### Simulate Poisson Distribution
