### Define Set Criteria
A subjects include patients with BMI < 26. B subjects contains patients older than 45.

```r
index_BMI_less_26 = BMI < 26 # Criteria for A
index_BMI_less_26
```

```r
index_AGE_greater_45 = AGE > 45 # Criteria for B
index_AGE_greater_45
```

```r
A_set <- PID[index_BMI_less_26] # A
A_set
```

```r
B_set <- PID[index_AGE_greater_45] # B
B_set
```

```r
index_A_int_B_compl <- !(index_BMI_less_26 & index_AGE_greater_45)
```

```r
PID[index_A_int_B_compl]
```

```r
index_Ac_uni_Bc_compl <- (!index_BMI_less_26 | !index_AGE_greater_45)
```
